```python
# Given a linked list, swap every two adjacent nodes and return its head. You must solve the problem without modifying the values in the list's nodes (i.e., only nodes themselves may be changed.)
"""Approach: Three pointers for ease"""
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
class Solution:
    def swapPairs(self, head:ListNode) -> ListNode:
        if(not head):
            return None
        if( not head.next):
            return head
        dummy = ListNode(0)
        dummy.next = head
        prev = dummy
        curr = head
        while(curr and curr.next):
            # saving the left linkedlist before swap
            next = curr.next
            # swapping make a diagram
            curr.next = next.next
            next.next = curr
            prev.next = next
            # move pointers prev and curr no need to move next moves automatically due to 1st line 
            prev = curr 
            curr = curr.next
        return dummy.next
```

```python
# reverse the k nodes in a linked list
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
class Solution:
    def reverseKGroup(self, head: ListNode, k: int) -> ListNode:
        def leng(curr):
            if(not curr):
                return 0
            count = 1
            while(curr.next):
                count += 1
                curr = curr.next
            return count
        def rev_ll(curr,k):
            if(not curr):
                return curr
            prev = None
            curr = curr
            for _ in range(k):
                next = curr.next
                curr.next = prev
                prev = curr
                curr = next
            return prev
        if(k == 1):
            return head
        dummy = ListNode(0)
        dummy.next = head
        prev = dummy
        curr = head
        totalLength = leng(curr)
        while(totalLength >= k):
            tail = curr #previous elements before curr
            for _ in range(k - 1):
                tail = tail.next
            next_head = tail.next#later elements after curr
            rev = rev_ll(curr,k)
            prev.next = rev#attach back
            curr.next = next_head
            prev = curr 
            curr = next_head
            totalLength -= k
        return dummy.next
```
