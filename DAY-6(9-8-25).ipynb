```python
#Problem of the day
class Solution:
    def isPowerOfTwo(self, n: int) -> bool:
        if(n == 0):
            return False
        if(n == 1):
            return True
        while(n != 1):
            if(n % 2 != 0):
                return False
            n = n//2
        return True
```

```python
# Merge two ll sortde ones
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
class Solution:
    def mergeTwoLists(self, list1, list2):
        ## without using a new linked list
        dummy = ListNode()
        tail = dummy
        while(list1 and list2):
            if(list1.val < list2.val):
                tail.next = list1
                list1 = list1.next
            else:
                tail.next = list2
                list2 = list2.next
            tail = tail.next
        tail.next = list1 or list2
        return dummy.next
```

```python
# Merge k linked lists 
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
class Solution:
    def mergeKLists(self, lists):
        def merge_two_list(ans,list2):
            dummy = ListNode()
            tail = dummy
            while(ans and list2):
                if(ans.val < list2.val):
                    tail.next = ans 
                    ans = ans.next
                else:
                    tail.next = list2
                    list2 = list2.next
                tail = tail.next
            tail.next = ans or list2
            return dummy.next
        new_head = None
        for i in lists:
            new_head = merge_two_list(new_head,i)
        return new_head
```
