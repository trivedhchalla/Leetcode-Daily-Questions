```python
# Given root of Binary tree check if its balanced or not
"""
Logic: balanced if |left_height - right_heigh| <= 1 i.e diff bw level of left and right node can -1/0/1
"""
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
class Solution:
    def isBalanced(self, root:TreeNode) -> bool:
        def height(root):
            if root is None:
                return 0,True
            left_height,left_balance = height(root.left)
            right_height,right_balance = height(root.right)
            curr_height = 1 + max(left_height,right_height)
            is_balance = (left_balance and right_balance and abs(left_height - right_height)<=1)
            return curr_height,is_balance
        he,balanced = height(root)
        return balanced
```

```python
# Letter combination of Phone number
class Solution:
    def letterCombinations(self, digits: str) -> list[str]:
        from itertools import product
        if(len(digits) == 0):
            return []
        char = 'a'
        dct = {}
        for i in range(2,10):
            if(i != 7 and i != 9):
                dct[str(i)] = [char,chr(ord(char)+1),chr(ord(char)+2)]
                char = chr(ord(char)+3) 
            else:
                dct[str(i)] = [char,chr(ord(char)+1),chr(ord(char)+2),chr(ord(char)+3)]
                char = chr(ord(char)+4) 
        if(len(digits) == 1):
            return dct[digits]
        if(len(digits) == 2):
            arr1 = dct[digits[0]]
            arr2 = dct[digits[1]]
            return [''.join(chars) for chars in product(arr1,arr2)]
        if(len(digits) == 3):
            arr1 = dct[digits[0]]
            arr2 = dct[digits[1]]
            arr3 = dct[digits[2]]
            return [''.join(chars) for chars in product(arr1,arr2,arr3)]
        if(len(digits) == 4):
            arr1 = dct[digits[0]]
            arr2 = dct[digits[1]]
            arr3 = dct[digits[2]]
            arr4 = dct[digits[3]]
            return [''.join(chars) for chars in product(arr1,arr2,arr3,arr4)]
```

```python
# Cycle in Linked List return node where cycle start
class ListNode:
    def __init__(self, x):
        self.val = x
        self.next = None
class Solution:
    def detectCycle(self, head:ListNode) -> ListNode:
        slow = head
        fast = head
        arr = []
        while(fast is not None and fast.next is not None):
            fast = fast.next.next
            slow = slow.next
            if(fast == slow):
                slow = head
                while(slow != fast):
                    slow = slow.next
                    fast = fast.next
                return slow
        return None
```
